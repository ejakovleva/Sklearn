#### Задание 1

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
X = pd.DataFrame(boston.data, columns=boston.feature_names)
```

```python
X.head(10)
```

```python
y = pd.DataFrame(boston.target, columns=['price'])
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```

#### Задание 2

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
```

```python
model = RandomForestRegressor(n_estimators=1000, max_depth=12, random_state=42, n_jobs=6)
```

```python
model.fit(X_train, y_train.values[:, 0])
```

```python
y_pred=model.predict(X_test)
```

```python
r2_score(y_test, y_pred)
```

лучше работает модель RandomForest, поскольку у этой модели R2 выше, чем у модели с линейной регрессией

#### Задание 4

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
```

```python
DATASET_PATH = './creditcard.csv'
pd.read_csv
df = pd.read_csv(DATASET_PATH, sep=",")
```

```python
df.head()
```

```python
df['Class'].value_counts(normalize=True)
```

```python
df.info()
```

```python
pd.options.display.max_columns = 100
```

```python
df.head(10)
```

```python
X = df.drop(['Class'], axis = 1)
X.head()
```

```python
y = pd.Series(df.loc[:, 'Class'])
y.head(100)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
parameters = {
    'n_estimators': [10, 15],
    'max_features': np.arange(3, 5),
    'max_depth': np.arange(4, 7),
}

clf = GridSearchCV(
        estimator=RandomForestClassifier(random_state=100),
        param_grid=parameters,
        scoring='roc_auc',
        cv=3,
        n_jobs=6)
```

```python
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
clf = RandomForestClassifier(max_depth=6, max_features=3, n_estimators=15)
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
y_pred_proba=clf.predict_proba(X_test)
```

```python
y_pred_proba=y_pred_proba[:, 1]
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(y_test, y_pred_proba)
```

```python
y_pred_proba_train=clf.predict_proba(X_train)
y_pred_proba_train=y_pred_proba_train[:, 1]
```

```python
roc_auc_score(y_train, y_pred_proba_train)
```

```python
from sklearn.metrics import roc_curve
from pylab import rcParams
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba, pos_label = 1)
```

```python
rcParams['figure.figsize'] = 5, 4.5
plt.plot(fpr, tpr, color='green', label='ROC curve')
plt.plot([0, 1], [0, 1], color='grey', linestyle='--')
plt.xlim(0.0, 1.0)
plt.ylim(0.0, 1.0)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating characteristic (ROC)')
```
